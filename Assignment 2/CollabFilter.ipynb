## Formatting the ratings matrix to be one row per user and one column per movie

```python
R_df = ratings_df.pivot(index = 'movieId', columns ='userId', values = 'rating').fillna(0)
# R_df.head()
display(R_df)
```

### Normalize by each users mean and convert it from a dataframe to a numpy array

```python
R = R_df.to_numpy()
user_ratings_mean = np.mean(R, axis = 1)
R_demeaned = R - user_ratings_mean.reshape(-1, 1)
```

```python
display(R_demeaned.shape)
display(R_demeaned)
# display(user_ratings_mean)
# display(R_demeaned[0])
# display(R_demeaned[1])
```

```python
user_ratings_pd_mean = pd.DataFrame.mean(R_df, axis=1)
# user_ratings_pd_mean = pd.DataFrame.mean(R_df, axis=1).to_frame("mean")["mean"]
# us = user_ratings_pd_mean.DataFrame.rename(columns={"": "mean"})
# us = user_ratings_pd_mean.columns.values[1] = 'mean'
# us = user_ratings_pd_mean.name = 'mean'
# us = user_ratings_pd_mean.rename(columns={"movieid": "movie", "": "mean"})
# us = user_ratings_pd_mean.columns.tolist()
# us = user_ratings_pd_mean.dtypes
display(user_ratings_pd_mean)

# us = user_ratings_pd_mean.stack()
# display(us)
```

```python
R_pd_demeaned = R_df.subtract(user_ratings_pd_mean, axis='index', fill_value=None)
display(R_pd_demeaned)
```
